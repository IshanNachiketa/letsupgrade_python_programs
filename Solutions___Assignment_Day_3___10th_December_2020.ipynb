 **Question 1 : Use IF ELSE and ELIF to write a program in python for your Report Cards.**

```
#Solution 1
 
print('Student report card')
print('=======================================')
stud_name=input('Please enter Name :')
stud_age=int(input('Pleae enter Age :'))
english_marks=int(input('Please enter English marks :'))
maths_marks=int(input('Please enter Maths marks :'))
science_marks=int(input('Please enter Science marks :'))
sst_marks=int(input('Please enter Sst marks :'))
computer_marks=int(input('Please enter Computer marks :'))
#Student totals and percentage
stud_total_marks = maths_marks + science_marks + english_marks + sst_marks + computer_marks
stud_percentage = round(stud_total_marks/500 * 100,2)
#Print student Report card
print('=======================================')
print(f'English       : {english_marks}')
print(f'Maths         : {maths_marks}')
print(f'Science       : {science_marks}')
print(f'Sst           : {sst_marks}')
print(f'Computer      : {computer_marks}')
print(f'Total         : {stud_total_marks}')
print(f'Percentage    : {stud_percentage}%')
print('=======================================')
#Student marks classification
if(stud_percentage>=75): 
  print('Grade : Distinction')
elif stud_percentage >=60 and  stud_percentage< 75:
  print('Grade : Very Good')
elif (stud_percentage>35 and stud_percentage<60):
  print('Grade : Good')
else:
  print('Fail')
print('=======================================')
```

**Question 2 : Use For Loop to Print Prime Numbers in between 1 to 1000**

```
#Solution 2
 
for i in range(1,1001):
  if i>1:
    for num in range(2,i):
      if (i%num==0):
        break
    else:  
        print(i)
```

 **Question 3 : Write a program for printing the tables from 1,10 using Nested For Loop.**

```
#Solution 3
 
# Multiplication table of 1 to 10
 
for i in range(1,11):
    print("\n\nMULTIPLICATION TABLE FOR %d\n" %(i))
    for j in range(1,11):
        print("%-5d X %5d = %5d" % (i, j, i*j))
```

 **Question 4 : Write a program to Print X Prime Numbers using For Loop starting from 0, and take the INput of X from
the user.**

```
#Solution 4
 
x=int(input('Please enter X for upper limit :'))
start = 0
end = x
 
for i in range(start,end):
  if i>1:
    for j in range(2,i):
        if(i % j==0):
            break
    else:
        print(i)
```
